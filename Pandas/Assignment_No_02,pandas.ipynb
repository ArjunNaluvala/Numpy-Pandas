#import the titanic dataset

```python
titanic_data = pd.read_csv('titanic.csv')

titanic_df = pd.DataFrame(titanic_data)
titanic_df
```

##### 1. Write a Pandas program to create a Pivot table and find the total fare amount class wise, gender wise.

```python
#### code here

pd.pivot_table(titanic_df,values='fare',index=['class','sex'],aggfunc = 'sum')
```

##### 2. Write a Pandas program to create a Pivot table and find survival rate by gender.

```python
### code here

pd.pivot_table(titanic_df,values='survived',index='sex',aggfunc='mean')
```

###### 3. Write a Pandas to create to find the total count no of people survived pclass wise, class wise and gender wise.

```python
### code here

pd.pivot_table(
    titanic_df,
    values='survived',
    index=['pclass','class','sex'],
    aggfunc='sum'
    )
```

###### 4. Write a Pandas program to create a Pivot table and calculate how many women and men were in a particular cabin class.

```python
###  code here

pd.pivot_table(
    titanic_df,
    index='class',
    columns='who',
    aggfunc='size'
)
```

##### 6. Write a Pandas program to replace null values with the value from the previous row or the next row in a given DataFrame.

```python
### code here

#backward filling i.e. filling with the next known value
#titanic_df['deck'] = titanic_df['deck'].fillna(method="bfill")

#forward filling i.e. filling with the last known value
#titanic_df['deck'] = titanic_df['deck'].fillna(method="ffill")

#Backward fill after forward fill

titanic_df['deck'] = titanic_df['deck'].ffill().bfill()
```

```python
titanic_df['deck'].isnull().sum()
```

###### 7. Write a Pandas program to replace nulls, with median or mean of the specified columns in a given DataFrame.

```python
### code here

#filling the null values in 'age' column with median

titanic_df['age'] = titanic_df['age'].fillna(titanic_df['age'].median())
```

```python
titanic_df['age'].isnull().sum()
```
