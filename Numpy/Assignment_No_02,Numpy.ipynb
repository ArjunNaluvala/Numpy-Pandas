```python
#Vectorization vs Loops
```

```python
arr = np.array([1,2,3,4,5])

print('Using a for loop:')
for i in arr:
    sq = i**2
    print(sq,end = ' ')
print(end='\n')

print('Using NumPy Vectorization:')
arr1 = arr**2
print(arr1)

print('NumPy Vectorization takes less number of lines and easier code to write.')
```

```python
arr = np.random.randint(0,11,10)
print('Array:',arr)

count=0
for i in arr:
    if i>5:
        count += 1
    else:
        pass
print('Count using for loop:',count)

count_vec = np.sum(arr > 5)
print("Count using vectorization:", count_vec)
```
